▲ 程序清单 1.43 if 语句和 elif 代码块的条件分支

```python
import random
num = random.randint(-10, 10)

if num > 0:
    print(num, '为正的数')
elif num == 0:
    print(num, '为零')
else:
    print(num, '为负的数')
```

▲ 程序清单 1.44 if 语句、elif 代码块及 else 代码块的条件分支

```python
n = 5
total = 0
i = 1

while i <= n:
    total = total + i
    print(i, '次合计 :', total )
    i = i + 1
```

▲  程序清单 1.54 在定义后的 multi 函数中指定参数 3 和参数 5 并执行处理

```python
def absolute(a):
    if a >= 0:
        return a
    else:
        return -1 * a

print(absolute(5))
print(absolute(-5))
```
